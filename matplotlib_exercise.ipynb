## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist_data = pd.read_csv("data/hist_data.csv")
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.hist(df_hist_data["aデータ"], bins=30, color="y", alpha=0.5, label="aデータ")
plt.hist(df_hist_data["bデータ"], bins=30, color="c", alpha=0.5, label="bデータ")
plt.legend(loc="upper right")
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter_data = pd.read_csv("data/scatter_data.csv")
data0 = df_scatter_data["class"]==0
data1 = df_scatter_data["class"]==1
data2 = df_scatter_data["class"]==2
plt.scatter(df_scatter_data[data0]["length (cm)"], df_scatter_data[data0]["width (cm)"], color="red", label="class0")
plt.scatter(df_scatter_data[data1]["length (cm)"], df_scatter_data[data1]["width (cm)"], color="blue", label="class1")
plt.scatter(df_scatter_data[data2]["length (cm)"], df_scatter_data[data2]["width (cm)"], color="green", label="class2")
plt.title("グラフ0〜2のlengthとwidthに関する散布図")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.grid(True)
plt.legend(loc="upper right")
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line_data = pd.read_csv("data/line_data.csv")
#折れ線グラフ
plt.plot(df_line_data["期間"], df_line_data["件数"], marker='o')
plt.title("期間ごとの件数の推移")
plt.xlabel("期間")
plt.ylabel("件数")
plt.grid(True)
plt.xticks(rotation=30)
plt.show()

#棒グラフ
plt.bar(df_line_data["期間"], df_line_data["費用"])
plt.title("期間ごとの費用の推移")
plt.xlabel("期間")
plt.ylabel("費用(円)")
plt.grid(True)
plt.xticks(rotation=30)
plt.show()
```
